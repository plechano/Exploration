```python
# VARIANTE PAULINE pour ne lire que 2012 à 2016 et pas tous les .csv

# les fichiers ETIQ2012-2013-2014-2015 sont de type .csv nativement
annees = range(2012, 2016)
fichiers_csv = [f"ETIQ{annee}.csv" for annee in annees]
print("Liste des fichiers.csv natifs avec la meme structure",fichiers_csv)
```

```python
#UnicodeDecodeError: 'utf-8' codec can't decode byte 0x82 in position 26204: invalid start byte
#celà signifie qu'il y a un problème d'encodage pour la lecture de nos fichiers csv
# Utilisation de chardet (avec chardet.detect()) qui est un des outils de détection automatique pour déterminer l'encodage en analysant son contenu.

# Création d'une boucle pour les lire, afficher les premières lignes, voire les infos

import chardet
dfs_ETIQ = {}

# Chargement des fichiers CSV
for fichier_csv in fichiers_csv:
    chemin_fichier = os.path.join(path, fichier_csv)
    
    # Utilisation de chardet à l'intérieur de la boucle pour détecter l'encodage du chaque fichier
    with open(chemin_fichier, 'rb') as f:
        result = chardet.detect(f.read())
    
    # Application de l'encodage détecté
    df_ETIQ = pd.read_table(chemin_fichier, sep=None, engine='python', encoding=result['encoding'])

    dfs_ETIQ[fichier_csv] = df_ETIQ

# Affichage des df_ETIQ crées
for nom_fichier, df in dfs_ETIQ.items():
    print("Nom du fichier:", nom_fichier)
    display(df_ETIQ.head(2))
```
